# Keyboard shortcuts

**The Jupyter Notebook with jupyter-vim-binding has two different keyboard input modes.Jupyter mode binds the keyboard to notebook level commands and is indicated by a grey cell border with a blue left margin.Vim command mode allows you to type code/text into a cell and is indicated by a green cell border.**

## Mac OS X modifier keys:

- ⌘: Command
- ⌃: Control
- ⌥: Option
- ⇧: Shift
- ↩: Return
- ␣: Space
- ⇥: Tab
- Jupyter Mode (press Shift-Esc to enable)
- /: find and replace
- ⌃D: scroll notebook down
- ⌃E: scroll down
- ⌃⇧D: scroll notebook down
- ⌃⇧U: scroll notebook up
- ⌃U: scroll notebook up
- ⌃Y: scroll up
- G,G: select the first cell
- ⇧G: select the last cell
- Z,M: collapse output
- Z,R: expand output
- Z,⇧M: collapse all output
- Z,⇧R: expand all output
- ↩: enter edit mode
- I: enter edit mode
- ⌘⇧P: open the command palette
- ⇧↩: run cell, select below
- ⌘↩: run selected cells
- ⌥↩: run cell and insert below
- ⌘⇧↩: run all cells
- ⌘1: change cell to code
- ⌘2: change cell to markdown
- ⌘3: change cell to raw
- 1: change cell to heading 1
- 2: change cell to heading 2
- 3: change cell to heading 3
- 4: change cell to heading 4
- 5: change cell to heading 5
- 6: change cell to heading 6
- ⌃K: select cell above
- K: select cell above
- ⌃J: select cell below
- J: select cell below
- ⌃⇧K: extend selected cells above
- ⇧K: extend selected cells above
- ⌃⇧J: extend selected cells below
- ⇧J: extend selected cells below
- Minus: split cell at cursor
- Subtract: split cell at cursor
- ⇧O: insert cell above
- O: insert cell below
- D,D: cut selected cells
- Y,Y: copy selected cells
- ⇧P: paste cells above
- P: paste cells below
- U: undo cell deletion
- ⇧M: merge selected cells, or current cell with cell below if only one cell is selected
- ⌘S: Save and Checkpoint
- ⇧L: toggle line numbers
- ⇧V: toggle output of selected cells
- Z,A: toggle output of selected cells
- ⇧S: toggle output scrolling of selected cells
- F1: show keyboard shortcuts
- ⇧H: show keyboard shortcuts
- 0,0: restart the kernel (with dialog)
- Esc: close the pager
- Q: close the pager
- ⇧R: Rename the current notebook
- Z,⇧A: Toggle the hidden state of all output areas
- Z,T: Scroll the current cell to the top
- Z,Z: Scroll the current cell to the center
